```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for column in df.columns:
    df[column] = LE.fit_transform(df[column])
```

```python
df.info()
```

```python
final_res = []
```

```python
x = df.drop('diagnosed_diabetes', axis =1)
y = df['diagnosed_diabetes']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
y_pred_LR = LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Logistic Regression :')
print('Accuracy Score',accuracy_score(y_test,y_pred_LR))
acc_LR = accuracy_score(y_test,y_pred_LR)
final_res.append(acc_LR)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train,y_train)
```

```python
y_pred_RF = RF.predict(x_test)
```

```python
print('Random Forest Classifier :')
print('Accuracy Score ',accuracy_score(y_test,y_pred_RF))
acc_RF = accuracy_score(y_test,y_pred_RF)
final_res.append(acc_RF)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
y_pred_DT = DT.predict(x_test)
```

```python
print('Decision Tree :')
print('Accuracy Score ',accuracy_score(y_test,y_pred_DT))
acc_DT = accuracy_score(y_test,y_pred_DT)
final_res.append(acc_DT)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(x_train,y_train)
```

```python
y_pred_KNN = KNN.predict(x_test)
```

```python
print('K Nearest Neighbors ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_KNN))
acc_KNN = accuracy_score(y_test,y_pred_KNN)
final_res.append(acc_KNN)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier()
GBC.fit(x_train,y_train)
```

```python
y_pred_GBC = GBC.predict(x_test)
```

```python
print('Gradient Boosting ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_GBC))
acc_GBC = accuracy_score(y_test,y_pred_GBC)
final_res.append(acc_GBC)
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier()
XGB.fit(x_train,y_train)
```

```python
y_pred_XGB = XGB.predict(x_test)
```

```python
print('XGBoost ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_XGB))
acc_XGB = accuracy_score(y_test,y_pred_XGB)
final_res.append(acc_XGB)
```

```python
plt.xticks(rotation = 90)
sns.barplot(x = ['Logistic Regression', 'Random Forest','Decision Tree','K-Nearest Neighbors','GradientBoosting','XGBoost'], y = 
            [accuracy_score(y_test,y_pred_LR),accuracy_score(y_test,y_pred_RF),
             accuracy_score(y_test,y_pred_DT),accuracy_score(y_test,y_pred_KNN),
             accuracy_score(y_test,y_pred_GBC),accuracy_score(y_test,y_pred_XGB)])
```

```python
final = np.array(final_res)
result = final.reshape(-1,1)
columns=['Accuracy']
index = ['Logistic Regression', 'Random Forest','Decision Tree','K-Nearest Neighbors','GradientBoosting','XGBoost']
final_result = pd.DataFrame(result,index = index,columns = columns)
final_result.sort_values(by='Accuracy', ascending=False)
```

## HyperParameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
grid_rf = RandomForestClassifier(n_estimators=1100,min_samples_leaf=2,min_samples_split=5,max_features="sqrt",max_depth=35)
grid_rf.fit(x_train,y_train)
```

```python
y_pred_grid_rf = grid_rf.predict(x_test)
```

```python
print('Accuracy Score: ', accuracy_score(y_test, y_pred_grid_rf))
acc_score_rf = accuracy_score(y_test, y_pred_grid_rf)
```

```python
grid_gb = GradientBoostingClassifier(
    n_estimators=300,         # number of boosting stages (trees)
    learning_rate=0.05,       # controls contribution of each tree
    max_depth=4,              # depth of individual trees
    min_samples_split=5,      # minimum samples to split node
    min_samples_leaf=3,       # minimum samples in leaf node
    subsample=0.9,            # fraction of samples used for fitting
    random_state=42
)

# Train the tuned model
grid_gb.fit(x_train, y_train)
```

```python
y_pred_grid_gb = grid_gb.predict(x_test)
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred_grid_gb))
acc_score_gb = accuracy_score(y_test, y_pred_grid_gb)
```

```python
models_score = pd.DataFrame({'Accuracy Score':[acc_score_rf,acc_score_gb]}, index=['Random forest','Gradient Boosting'])
models_score.sort_values(by = 'Accuracy Score',ascending = False)
```
